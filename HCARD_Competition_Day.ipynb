```python
#@title Game 1 - CANS

# time_convert function to convert time into hours, minutes and seconds
def time_convert(sec):
  mins = sec // 60
  sec = sec % 60
  hours = mins // 60
  mins = mins % 60
  print("Time Lapsed = {0}".format(sec))


# Starting the game
game_on = True
win_lose = ''
game_over = False

while game_on:

  # Generate the secret code 
  secret_code = [random.randint(0,9), random.randint(0,9), random.randint(0,9), random.randint(0,9)]
  answer = False

  print(cans[secret_code[0]], cans[secret_code[1]], cans[secret_code[2]], cans[secret_code[3]])


  print("Enter your username!")
  name = input()
  print("\nFind a 4-number Secret Code.\nYou will have a max of {0} seconds!\nWhen ready, click on 'Enter'!".format(exceeded_time))
  input()
  start = time.time()

  # Keep playing if the answer is not found
  while not answer:
    start_time = time.time()

    print('\nEnter the secret code:')
    x = input()

    # If code entered is not numbers and player still has time to play
    if (not x.isdigit() and time.time() - start < exceeded_time):
      print("The Secret Code must be 4 numbers!\n")

    # If time is exceeded
    elif (time.time() - start > exceeded_time):
      display(Audio(path + 'sound.wav', autoplay=True))
      answer = True
      time_lapsed = exceeded_time
      win_lose = 'Lose'
      game_over = True
      print("\nTime Exceeded \nThe Secret Code was: {0}{1}{2}{3}".format(secret_code[0], secret_code[1], secret_code[2], secret_code[3]))

    # If code length is not corresponding
    elif (not x.isdigit() or not (len(x) == 4)):
      print("The Secret Code must be 4 numbers!\n")

    # If secret code is found in time
    elif (int(x[0]) == secret_code[0] and int(x[1]) == secret_code[1] and int(x[2]) == secret_code[2] and int(x[3]) == secret_code[3]):
      answer = True
      print("\nCongratulations! " + x + " was the Secret Code\n")
      end_time = time.time()
      time_lapsed = end_time - start_time
      time_convert(time_lapsed)
      win_lose = 'Win'
      game_over = True

    # If secret code is still not found
    else:
      print("You have at least one mistake!\n")



    # Game Over
    if (game_over):
      # Update Data Sheet and Leaderboard
      data1.update_acell('A' + str(test_number_1 + 1), test_number_1)
      data1.update_acell('B' + str(test_number_1 + 1), name)
      data1.update_acell('C' + str(test_number_1 + 1), "{0}{1}{2}{3}".format(secret_code[0], secret_code[1], secret_code[2], secret_code[3]))
      data1.update_acell('D' + str(test_number_1 + 1), win_lose)
      data1.update_acell('E' + str(test_number_1 + 1), time_lapsed)  

      leaderboard.format('A2:D1000', {"backgroundColor": {"red": 1.0, "green": 1.0, "blue": 1.0}})
      leaderboard.update_acell('B' + str(test_number_1 + 1), name)
      leaderboard.update_acell('C' + str(test_number_1 + 1), time_lapsed)
      leaderboard.update_acell('D' + str(test_number_1 + 1), '= row() - 1')  
      leaderboard.format('A' + str(test_number_1 + 1) + ':D' + str(test_number_1 + 1), {"backgroundColor": {"red": 0.0, "green": 0.9, "blue": 0.0}})
      leaderboard.sort((3, 'asc'), range='A2:D1000')
      
      # Update test number
      test_number_1 += 1

      # Restart a Game
      print("\n\n\n\n\nStart a new Game? (y/n)\n")
      new_game = input()

      if not (new_game == 'y' or new_game == 'Y'):
        game_on = False
      else:
        game_over = False

      output.clear()    
```
