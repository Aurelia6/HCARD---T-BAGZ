```python
#@title All Functions

# time_convert function to convert time into hours, minutes and seconds
def time_convert(sec):
  mins = sec // 60
  sec = sec % 60
  hours = mins // 60
  mins = int(mins % 60)
  print("Time Lapsed = {0} min {1} seconds".format(mins, sec))

def game1(name, test_number, max_time, cans, number):
  output.clear()
  # Starting the game
  win_lose = ''
  game_over = False

  while True:

    # Generate the secret code 
    secret_code = secret_code = random.sample(range(len(cans)), number)
    answer = False

    # print(secret_code)
    input("{0}\nFind a {1}-number Secret Code.\nYou will have a maximum of {2} seconds!\nWhen ready, click on 'Enter'!\n".format(game1_name, number, max_time))
    if (number == 2):
      print("This is the order you have to follow:\n {0} - {1}\n".format(cans[secret_code[0]], cans[secret_code[1]]))
    elif (number == 3):
      print("This is the order you have to follow:\n {0} - {1} - {2}\n".format(cans[secret_code[0]], cans[secret_code[1]], cans[secret_code[2]]))
    else:
      print("This is the order you have to follow:\n {0} - {1} - {2} - {3}\n".format(cans[secret_code[0]], cans[secret_code[1]], cans[secret_code[2]], cans[secret_code[3]]))

    start = time.time()

    # Keep playing if the answer is not found
    while not answer:
      start_time = time.time()

      x = input("\nEnter the secret code:\n")

      # If code entered is not numbers and player still has time to play
      if (not x.isdigit() and time.time() - start < max_time):
        print("The Secret Code must be {0} numbers!\n".format(number))

      # If time is exceeded
      elif (time.time() - start > max_time):
        display(Audio(path + 'fail.wav', autoplay=True))
        time.sleep(3)
        output.clear()
        answer = True
        time_lapsed = max_time
        win_lose = 'Lose'
        game_over = True
        if (number == 2):
          print("Time Exceeded \nThe Secret Code was: {0}{1}".format(secret_code[0], secret_code[1]))
        elif (number == 3):
          print("Time Exceeded \nThe Secret Code was: {0}{1}{2}".format(secret_code[0], secret_code[1], secret_code[2]))
        else:
          print("Time Exceeded \nThe Secret Code was: {0}{1}{2}{3}".format(secret_code[0], secret_code[1], secret_code[2], secret_code[3]))

      # If code length is not corresponding
      elif (not x.isdigit() or not (len(x) == number)):
        print("The Secret Code must be {0} numbers!\n".format(number))

      # If secret code is found in time
      elif ((number == 4 and int(x[0]) == secret_code[0] and int(x[1]) == secret_code[1] and int(x[2]) == secret_code[2] and int(x[3]) == secret_code[3]) or (number == 3 and int(x[0]) == secret_code[0] and int(x[1]) == secret_code[1] and int(x[2]) == secret_code[2]) or (number == 2 and int(x[0]) == secret_code[0] and int(x[1]) == secret_code[1])):
        answer = True
        end_time = time.time()
        time_lapsed = end_time - start_time
        win_lose = 'Win'
        game_over = True

        display(Audio(path + 'success.mp3', autoplay=True))
        time.sleep(5)
        output.clear()

        print("Congratulations! " + x + " was the Secret Code")
        time_convert(time_lapsed)

      # If secret code is still not found
      else:
        print("You have at least one mistake!\n")



      # Game Over
      if (game_over):

        # Update Data Sheet and Leaderboard
        data1.update_acell('A' + str(test_number + 1), test_number)
        data1.update_acell('B' + str(test_number + 1), name)
        if (number == 2):
          data1.update_acell('C' + str(test_number + 1), "{0}{1}".format(secret_code[0], secret_code[1]))
        elif (number == 3):
          data1.update_acell('C' + str(test_number + 1), "{0}{1}{2}".format(secret_code[0], secret_code[1], secret_code[2]))
        else:
          data1.update_acell('C' + str(test_number + 1), "{0}{1}{2}{3}".format(secret_code[0], secret_code[1], secret_code[2], secret_code[3]))
        data1.update_acell('D' + str(test_number + 1), win_lose)
        data1.update_acell('E' + str(test_number + 1), time_lapsed)  

        leaderboard.format('B2:D1000', {"backgroundColor": {"red": 1.0, "green": 1.0, "blue": 1.0}})
        leaderboard.update_acell('B' + str(test_number + 1), '= row() - 1')  
        leaderboard.update_acell('C' + str(test_number + 1), name)
        leaderboard.update_acell('D' + str(test_number + 1), time_lapsed)

        leaderboard.format('B' + str(test_number + 1) + ':D' + str(test_number + 1), {"backgroundColor": {"red": 0.0, "green": 0.9, "blue": 0.0}})
        leaderboard.sort((4, 'asc'), range='B2:D1000')

        return time_lapsed
 
def game2(name, test_number):
  output.clear()
  input("{}\nYou will have to press 'Enter' to start and stop the timer.\nWhen ready, click on 'Enter'!\n".format(game2_name))

  # Start Timer
  start_time = time.time()

  input("\nTimer started\n")
  

  # Stop Timer
  end_time = time.time()
  time_lapsed = end_time - start_time

  # Pay Audio
  display(Audio(path + 'success.mp3', autoplay=True))
  time.sleep(5)
  output.clear()

  # Display Time
  time_convert(time_lapsed)


  # Update Data Sheet and Leaderboard
  data2.update_acell('A' + str(test_number + 1), test_number)
  data2.update_acell('B' + str(test_number + 1), name)
  data2.update_acell('C' + str(test_number + 1), time_lapsed)  

  leaderboard.format('G2:I1000', {"backgroundColor": {"red": 1.0, "green": 1.0, "blue": 1.0}})
  leaderboard.update_acell('G' + str(test_number + 1), '= row() - 1')  
  leaderboard.update_acell('H' + str(test_number + 1), name)
  leaderboard.update_acell('I' + str(test_number + 1), time_lapsed)

  leaderboard.format('G' + str(test_number + 1) + ':I' + str(test_number + 1), {"backgroundColor": {"red": 0.0, "green": 0.9, "blue": 0.0}})
  leaderboard.sort((9, 'asc'), range='G2:I1000')

  # Return Time
  return time_lapsed

def game3(name, test_number):
  output.clear()

  input("{}\nYou will have to press 'Enter' to start and stop the timer.\nWhen ready, click on 'Enter'!\n".format(game3_name))

  # Start Timer
  start_time = time.time()

  input("\nTimer started\n")

  # Stop Timer
  end_time = time.time()
  time_lapsed = end_time - start_time

  # Pay Audio
  display(Audio(path + 'success.mp3', autoplay=True))
  time.sleep(5)
  output.clear()

  # Display Time
  time_convert(time_lapsed)


  # Update Data Sheet and Leaderboard
  data3.update_acell('A' + str(test_number + 1), test_number)
  data3.update_acell('B' + str(test_number + 1), name)
  data3.update_acell('C' + str(test_number + 1), time_lapsed)  


  leaderboard.format('K2:N1000', {"backgroundColor": {"red": 1.0, "green": 1.0, "blue": 1.0}})
  leaderboard.update_acell('L' + str(test_number + 1), '= row() - 1')  
  leaderboard.update_acell('M' + str(test_number + 1), name)
  leaderboard.update_acell('N' + str(test_number + 1), time_lapsed)

  leaderboard.format('L' + str(test_number + 1) + ':N' + str(test_number + 1), {"backgroundColor": {"red": 0.0, "green": 0.9, "blue": 0.0}})
  leaderboard.sort((14, 'asc'), range='L2:N1000')
  
  # Return Time
  return time_lapsed

def displayMixedLeaderboard(name, total_time, final_ranking):

  ## Store useful Data
  # Get all the values from current leaderboard
  all_values = leaderboard.get_all_values()
  # Reset the value for each key to 0 - since the ordering may change with new participants
  final_ranking = final_ranking.fromkeys(final_ranking, 0)
  # List with all the total time and their sorted indexes, so we can rank from time when ranking is equal
  all_total_times.append(total_time)
  sorted_total_times = numpy.argsort(numpy.argsort(all_total_times))

  ## Loop to get the sum of the total ranking based on each game rank
  for rank in range (1, len(final_ranking) + 1):
    for name in range(2, 13, 5): 
      final_ranking[str(all_values[rank][name])] =  int(final_ranking[str(all_values[rank][name])]) + rank

  ## Loop to define the rows in sheet, based on both the sum of rankings and time
  for index, (surname, ranking) in enumerate(final_ranking.items()):
    mixedLeaderboard.format('B2:D1000', {"backgroundColor": {"red": 1.0, "green": 1.0, "blue": 1.0}})
    mixedLeaderboard.format('B2:B1000',{"textFormat": {"foregroundColor": {"red": 1.0, "green": 1.0, "blue": 1.0}}})
    mixedLeaderboard.update_acell('B' + str(index + 2), int(ranking) * 100 + int(sorted_total_times[index]))
    mixedLeaderboard.update_acell('C' + str(index + 2), surname)
    mixedLeaderboard.update_acell('D' + str(index + 2), all_total_times[index])
  mixedLeaderboard.format('B' + str(test_number + 1) + ':D' + str(test_number + 1), {"backgroundColor": {"red": 0.0, "green": 0.9, "blue": 0.0}})
  mixedLeaderboard.sort((2, 'asc'), range='B2:D1000')

  ## Reset the ranking number for them to be in increasing order
  for index in range(test_number):
    mixedLeaderboard.update_acell('B' + str(index + 2), index + 1)
  mixedLeaderboard.format('B2:B1000',{"textFormat": {"foregroundColor": {"red": 0.0, "green": 0.0, "blue": 0.0}}})
```

```python
#@title Play All Games at Once
game_on = True

while game_on:
  # Setup vairable
  total_time = 0

  # Get Player Name and Store It
  name = input("You will play the 3 Games in the following order: \n - {0}\n - {1}\n - {2}\n\nEnter your Username!\n".format(game1_name, game2_name, game3_name))
  final_ranking[name] = 0

  # Call Functions To Play Games
  total_time += game1(name, test_number, max_time, cans, number)
  total_time += game2(name, test_number)
  total_time += game3(name, test_number)

  # Update the Total Leaderboard
  displayMixedLeaderboard(name, total_time, final_ranking)

  output.clear()
  test_number += 1

  # Restart a Game
  new_game = input("Thank You for Playing!\n\n\n\n\nStart a new Game? (y/n)\n")

  if not (new_game == 'y' or new_game == 'Y'):
    game_on = False

    
  output.clear()
```
